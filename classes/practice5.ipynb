```python
def calc_dists(pontos):
    n = len(pontos)
    dists = [[0 for i in range(n)] for j in range(n)]
    for i in range(n):
        for j in range(n):
            x1, y1 = pontos[i]
            x2, y2 = pontos[j]
            d = int(np.sqrt((x1 - x2)**2 + (y1 - y2)**2) + 0.5)
            dists[i][j] = d

    return dists
```

```python
def print_solution(model):
  print("Status = ", status)
  print("Solution value  = ", model.objective_value)

  print("Solution:")

  selected = []
  for v in model.vars:
      if v.x > 0.00001:
          print(v.name, " = ", v.x)
```

```python
def cria_modelo(n, R, dist): # implementar aqui
    model = Model(sense=MINIMIZE, solver_name=CBC)
    x = [model.add_var(var_type=BINARY, name=f"x_{i}") for i in range(n)] # Uma variável para cada distrito

    """
    - Será 0 se não tiver uma UPA naquele distrito, e 1 caso contrário
    """
    model.objective = sum(x) # Quero minimizar a quantidade de UPAs

    for i in range(n): # Para cada distrito
      soma = 0
      aux = list()
      for j in range(n):
        if dist[i][j] <= R: 
          aux.append(x[j])
      
      model += xsum(aux) >= 1
    model.write("model.lp")
    with open("model.lp") as file:
      print(file.read())
    return model
    
```
