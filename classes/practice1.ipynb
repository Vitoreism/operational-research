```python
from mip import *
```

**QUESTÃO 1**

```python
# Criando o modelo para o problema
modelo = Model(sense=MINIMIZE, solver_name=CBC)

# Criação das variáveis do problema

x = [modelo.add_var(var_type=CONTINUOUS, name=f"x_{i}", lb=0) for i in range(3)]

# Função objetivo -> Minimizar os custos para a criação da liga metálica
modelo.objective = 90*x[0] + 180*x[1] + 25*x[2]

# Restrições: 

modelo += x[0] + x[1] + x[2] == 10
modelo += 0.005*x[0] + 0.9 * x[1] + 0.09 * x[2] <= 0.95
modelo += 0.005*x[0] + 0.9 * x[1] + 0.09 * x[2] >= 0
modelo += 0.14 * x[0] + 0.27 * x[2] <= 2
modelo += 0.14 * x[0] + 0.27 * x[2] >= 1.9


modelo.write("modelo.lp")
with open("modelo.lp") as file:
    print(file.read())
```

```python
status = modelo.optimize()

print(f"Status = {status}")
print(f"Solution value = {modelo.objective_value:.4f}\n")

print("Solution: ")
for v in modelo.vars:
    print(f"{v.name} = {v.x:.2f}")
```
