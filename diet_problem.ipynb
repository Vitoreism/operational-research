```python
from mip import *
```

**Criação do modelo**

min1.5x0+2x1+3x2+0.8x3 
S.a.
7x0+8x1+30x2+6x3≥3 
550x0+300x1+400x2+250x3≥50 
x0,x1,x2,x3≥0

```python
# Criando o modelo para o problema
modelo = Model(sense=MINIMIZE, solver_name=CBC)

# Criação das variáveis do problema
x = [modelo.add_var(var_type=CONTINUOUS, name=f"x_{i}", lb=0) for i in range(4)]

# Função objetivo -> Minimizar o custo total
modelo.objective = 1.5*x[0] + 2*x[1] + 3*x[2] + 0.8*x[3]

# Restrição referente a quantidade mínima de vitamina A
modelo += 7*x[0] + 8*x[1] + 30*x[2] + 6*x[3] >= 3

# Restrição referente a quantidade mínima de vitamina C
modelo += 55*x[0] + 30*x[1] + 40*x[2] + 25*x[3] >= 5

modelo.write("modelo.lp")
with open("modelo.lp") as file:
    print(file.read())
```

```python
status = modelo.optimize()

print(f"Status = {status}")
print(f"Solution value = {modelo.objective_value:.4f}\n")

print("Solution: ")
for v in modelo.vars:
    print(f"{v.name} = {v.x:.4f}")
```
